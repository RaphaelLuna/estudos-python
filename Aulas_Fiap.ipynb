```python
from string import Template
nome, idade = 'Cibele', 30
print('Nome: %s Idade: %d' % (nome, idade)) # Mais antiga
print('Nome: {0} Idade: {1}'.format(nome,idade)) # python < 3.8
print(f'Nome: {nome} Idade: {idade}') # python >=3.6
s = Template ('Nome: $nome idade: $idade')
print(s.substitute(nome=nome, idade=idade))
```

```python
categoria = int(input("Digite a categoria do produto: "))
if categoria == 1:
  preco = 10
elif categoria == 2:
  preco = 18
elif categoria == 3:
  preco = 23
elif categoria == 4:
  preco = 26
elif categoria == 5:
  preco = 31
else:
  print("Categoria inválida")
  preco = 0
print(f"O preço do produto é: R${preco:6.2f}")
```

```python
idade = int(input("Digite sua idade: "))
while idade < 0:
  print("Idade preenchida errado, digite novamente!")
  idade = int(input("Digite uma nova idade: "))
if idade < 18:
  print('Menor de idade!')
elif idade in range (18, 65):
  print('Adulto')
elif idade in range (65, 100):
  print('melhor idade ')
elif idade > 100:
  print('Parabéns, você chegou lá!')

### Desta forma conseguimos realizar um programa de forma consistente e sem erros, caso algúem digite uma idade negativa
### Usamos o range para gerar um intervalo de numeros de x a y

```

```python
for i in range(1,11):
  print(i)
```

```python
for x in range(1, 11):
  for y in range(1, 11):
    print(f'{x} * {y} = {x*y}')
```
