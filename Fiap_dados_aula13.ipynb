```python
import numpy as np
import pandas as pd
```

```python
df_dados_paises = pd.read_csv('dadospaises.csv', sep=';', encoding='latin1', decimal=',')
```

```python
# estatísticas descritivas
df_dados_paises.groupby('pais')                .agg(min_idh = pd.NamedAgg('idh', 'min'),
                    max_idh = pd.NamedAgg('idh', 'max'),
                    media_idh = pd.NamedAgg('idh', 'mean'),
                    mediana_idh = pd.NamedAgg('idh', 'median'),
                    desvio_padrao_idh = pd.NamedAgg('idh', 'std'))
```

```python
# modulos
from scipy.stats import binom, poisson, norm
# binomial
n = 5
p = 0.8
binom.pmf(k=1, n = n, p = p)
# poisson
mu = 2
poisson.pmf(k=1, mu=mu)
# normal
media = 30
desvio_padrao = 6
norm.pdf(x = 25, loc = media, scale = desvio_padrao)
```
